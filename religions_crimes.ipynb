```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
religion_crimes = pd.read_csv("data")
```

```python
# Filter the dataset to include only the years ranging from 2000 to 2017
year_range_df = religion_crimes[(religion_crimes['Year'] >= 2000) & (religion_crimes['Year'] <= 2017)]
```

```python
year_range_df.sample(10)  
```

```python
# Define critical columns
critical_columns = ['Kidnappings per 100000 people', 'Robberies per 100000 people', 'Number of prisoners per 100000 people', 'Thefts per 100000 people', 'Homicides per 100000 people']

# Drop rows with null values in critical columns
filtered_df = year_range_df.dropna(subset=critical_columns)

# # Save the cleaned dataset to a new CSV file
# cleaned_file_path_2000_2017_specific = 'religious_crime/data'
# filtered_df_2000_2017_cleaned_specific.to_csv(cleaned_file_path_2000_2017_specific, index=False)

clean_df = filtered_df

clean_df. sample(10)
```

```python
# Load the cleaned dataset
clean_df = pd.read_csv('data')

# Get unique countries
unique_countries = clean_df['Country'].unique()

# Get summary statistics for religions
religions_summary = clean_df[[
    'Muslims as percent of the total population',
    'Christians as percent of the total population',
    'Non religious people as percent of the population',
    'People practicing Hinduism as percent of the population',
    'Buddhists as percent of the total population',
    'People practicing Judaism as percent of the population'
]].describe()

# Get summary statistics for crimes
crimes_summary = clean_df[[
    'Kidnappings per 100000 people',
    'Robberies per 100000 people',
    'Number of prisoners per 100000 people',
    'Thefts per 100000 people',
    'Homicides per 100000 people'
]].describe()

(unique_countries, religions_summary, crimes_summary)
```

```python
# Define thresholds for categorizing countries based on the percentage of Muslims
bins = [0, 20, 60, 100]
labels = ['Low', 'Medium', 'High']

# Categorize countries based on the percentage of Muslims
clean_df['Muslim_Category'] = pd.cut(clean_df['Muslims as percent of the total population'], bins=bins, labels=labels, include_lowest=True)

# Calculate average crime rates for each category
crime_columns = [
    'Kidnappings per 100000 people', 
    'Robberies per 100000 people', 
    'Number of prisoners per 100000 people', 
    'Thefts per 100000 people', 
    'Homicides per 100000 people'
]
average_crime_rates = clean_df.groupby('Muslim_Category')[crime_columns].mean()

# Display the average crime rates
print(average_crime_rates)
```

```python
# Plot the average crime rates
average_crime_rates.plot(kind='bar', figsize=(14, 8))
plt.title('Average Crime Rates by Muslim Category')
plt.ylabel('Crime Rate per 100,000 people')
plt.xlabel('Muslim Category')
plt.xticks(rotation=0)
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
bins_christians = [0, 20, 60, 100]
labels_christians = ['Low', 'Medium', 'High']

# Categorize countries based on the percentage of Christians
clean_df['Christian_Category'] = pd.cut(clean_df['Christians as percent of the total population'], bins=bins_christians, labels=labels_christians, include_lowest=True)

# Calculate average crime rates for each category
average_crime_rates_christians = clean_df.groupby('Christian_Category')[crime_columns].mean()

# Display the average crime rates for Christian categories
print(average_crime_rates_christians)
```
