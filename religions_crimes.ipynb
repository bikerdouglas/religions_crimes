```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
religion_crimes = pd.read_csv("religious_crime/religion_crime.csv")
```

```python
# Filter the dataset to include only the years ranging from 2000 to 2017
filtered_df_2000_2017 = religion_crimes[(religion_crimes['Year'] >= 2000) & (religion_crimes['Year'] <= 2017)]
```

```python
filtered_df_2000_2017.sample(10)  
```

```python
# Define critical columns
critical_columns = ['Kidnappings per 100000 people', 'Robberies per 100000 people', 'Number of prisoners per 100000 people', 'Thefts per 100000 people', 'Homicides per 100000 people']

# Drop rows with null values in the specified critical columns
filtered_df_2000_2017_cleaned_specific = filtered_df_2000_2017.dropna(subset=critical_columns)

# Save the cleaned dataset to a new CSV file
cleaned_file_path_2000_2017_specific = 'religious_crime/data'
filtered_df_2000_2017_cleaned_specific.to_csv(cleaned_file_path_2000_2017_specific, index=False)

cleaned_df = filtered_df_2000_2017_cleaned_specific

cleaned_df. sample(10)
```

```python
# Load the cleaned dataset
filtered_df_2000_2017_cleaned_specific = pd.read_csv('religious_crime/data')

# Get unique countries
unique_countries = filtered_df_2000_2017_cleaned_specific['Country'].unique()

# Get summary statistics for religions
religions_summary = filtered_df_2000_2017_cleaned_specific[[
    'Muslims as percent of the total population',
    'Christians as percent of the total population',
    'Non religious people as percent of the population',
    'People practicing Hinduism as percent of the population',
    'Buddhists as percent of the total population'
]].describe()

# Get summary statistics for crimes
crimes_summary = filtered_df_2000_2017_cleaned_specific[[
    'Kidnappings per 100000 people',
    'Robberies per 100000 people',
    'Number of prisoners per 100000 people',
    'Thefts per 100000 people',
    'Homicides per 100000 people'
]].describe()

(unique_countries, religions_summary, crimes_summary)
```

```python
# Define thresholds for categorizing countries based on the percentage of Muslims
bins = [0, 20, 60, 100]
labels = ['Low', 'Medium', 'High']

# Categorize countries based on the percentage of Muslims
cleaned_df['Muslim_Category'] = pd.cut(cleaned_df['Muslims as percent of the total population'], bins=bins, labels=labels, include_lowest=True)

# Calculate average crime rates for each category
crime_columns = [
    'Kidnappings per 100000 people', 
    'Robberies per 100000 people', 
    'Number of prisoners per 100000 people', 
    'Thefts per 100000 people', 
    'Homicides per 100000 people'
]
average_crime_rates = cleaned_df.groupby('Muslim_Category')[crime_columns].mean()

# Display the average crime rates
print(average_crime_rates)
```

```python
# Plot the average crime rates
average_crime_rates.plot(kind='bar', figsize=(14, 8))
plt.title('Average Crime Rates by Muslim Category')
plt.ylabel('Crime Rate per 100,000 people')
plt.xlabel('Muslim Category')
plt.xticks(rotation=0)
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Correct the reset and renaming of the index
average_crime_rates_all = pd.concat([
    average_crime_rates,
    average_crime_rates_christians,
    average_crime_rates_non_religious,
    average_crime_rates_hinduism,
    average_crime_rates_buddhism
], keys=['Muslims', 'Christians', 'Non-Religious', 'Hinduism', 'Buddhism'], names=['Religion', 'Religious_Category'])

# Reset index for better plotting
average_crime_rates_all = average_crime_rates_all.reset_index()

# Plot all crime columns with all religions on the same graph
fig, axes = plt.subplots(nrows=len(crime_columns), ncols=1, figsize=(14, 30))

for i, crime_column in enumerate(crime_columns):
    ax = axes[i]
    for religion, df_group in average_crime_rates_all.groupby('Religion'):
        df_group.plot(x='Religious_Category', y=crime_column, ax=ax, label=religion, marker='o')
    ax.set_title(f'Average {crime_column} by Religious Category')
    ax.set_xlabel('Religious Category')
    ax.set_ylabel(crime_column)
    ax.legend(title='Religion', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(True)

plt.tight_layout()
plt.show()
```
