```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
religion_crimes = pd.read_csv("data")
```

```python
null_counts = religion_crimes.isnull().sum()
null_counts
```

```python
critical_columns = [
    'Kidnappings per 100000 people', 
    'Robberies per 100000 people', 
    'Number of prisoners per 100000 people', 
    'Thefts per 100000 people', 
    'Homicides per 100000 people'
]

# Drop rows with null values only in the crime rate columns
clean_df = religion_crimes.dropna(subset=critical_columns)

# Display the first few rows of the cleaned dataframe
clean_df.tail(10)
```

```python
# Filter the data for the range of 2005-2017
filtered_data = clean_df[(clean_df['Year'] >= 2000) & (clean_df['Year'] <= 2017)]
filtered_data
```

```python
crime_columns=['Kidnappings per 100000 people', 'Robberies per 100000 people', 'Number of prisoners per 100000 people', 'Thefts per 100000 people', 'Homicides per 100000 people']
```

```python
# Define thresholds for categorizing countries based on the percentage of Muslims
bins = [0, 20, 60, 100]
labels = ['Low', 'Medium', 'High']

# Categorize countries based on the percentage of Muslims
filtered_data['Muslim_Category'] = pd.cut(filtered_data['Muslims as percent of the total population'], bins=bins, labels=labels, include_lowest=True)

# Calculate average crime rates for each category
average_crime_rates_m = filtered_data.groupby('Muslim_Category')[crime_columns].mean() # type: ignore
average_crime_rates_m
```

```python
# Plot the average crime rates
average_crime_rates_m.plot(kind='bar', figsize=(14, 8))
plt.title('Average Crime Rates by Practicing Muslims')
plt.ylabel('Crime Rate per 100,000 people')
plt.xlabel('Muslim Category')
plt.xticks(rotation=0)
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
average_crime_rates_all = pd.concat([
    average_crime_rates_m,
    average_crime_rates_c,
    average_crime_rates_non,
    average_crime_rates_h,
    average_crime_rates_b,
    average_crime_rates_j
], keys=['Muslims', 'Christians', 'Non-Religious', 'Hinduism', 'Buddhism', 'Judaism'], names=['Religion', 'Religious_Category'])

# Reset index for better plotting
average_crime_rates_all = average_crime_rates_all.reset_index()

# Plot all crime columns with all religions on the same graph
fig, axes = plt.subplots(nrows=len(crime_columns), ncols=1, figsize=(14, 30))

for i, crime_column in enumerate(crime_columns):
    ax = axes[i]
    for religion, df_group in average_crime_rates_all.groupby('Religion'):
        df_group.plot(x='Religious_Category', y=crime_column, ax=ax, label=religion, marker='o')
    ax.set_title(f'Average {crime_column} by Religious Category')
    ax.set_xlabel('Religious Category')
    ax.set_ylabel(crime_column)
    ax.legend(title='Religion', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# List of crime columns
crime_columns = [
    'Kidnappings per 100000 people', 
    'Robberies per 100000 people', 
    'Number of prisoners per 100000 people', 
    'Thefts per 100000 people', 
    'Homicides per 100000 people'
]

# List of religious composition columns
religion_columns = [
    'Christians as percent of the total population', 
    'Muslims as percent of the total population',
    'Buddhists as percent of the total population', 
    'People practicing Hinduism as percent of the population',
    'Non religious people as percent of the population',
    'People practicing Judaism as percent of the population'
]

# Combine crime and religious columns
all_columns = crime_columns + religion_columns

# Calculate average crime rates and religious composition for each country
average_data_by_country = filtered_data.groupby('Country')[all_columns].mean().reset_index()

# Plot each crime column on a world map using Plotly
for crime_column in crime_columns:
    fig = px.choropleth(average_data_by_country,
                        locations="Country",
                        locationmode='country names',
                        color=crime_column,
                        hover_name="Country",
                        hover_data=religion_columns,  # Add religious composition to hover data
                        title=f'Average {crime_column} by Country',
                        color_continuous_scale=px.colors.sequential.Hot_r)
    fig.show()
```

```python
# List of crime columns
crime_columns = [
    'Kidnappings per 100000 people', 
    'Robberies per 100000 people', 
    'Number of prisoners per 100000 people', 
    'Thefts per 100000 people', 
    'Homicides per 100000 people'
]

# List of religious composition columns
religion_columns = [
    'Christians as percent of the total population', 
    'Muslims as percent of the total population',
    'Buddhists as percent of the total population', 
    'People practicing Hinduism as percent of the population',
    'Non religious people as percent of the population',
    'People practicing Judaism as percent of the population'
]

# Combine crime and religious columns
all_columns = crime_columns + religion_columns

# Calculate the overall average crime rate for each country
filtered_data['Overall Average Crime Rate'] = filtered_data[crime_columns].mean(axis=1)

# Calculate the average crime rates and religious composition for each country
average_data_by_country = filtered_data.groupby('Country')[all_columns + ['Overall Average Crime Rate']].mean().reset_index()

# Aggregate data based on the religious composition
religion_aggregated_data = average_data_by_country[religion_columns + ['Overall Average Crime Rate']]

# Reshape the data to long format for easier comparison
long_format_data = pd.melt(religion_aggregated_data, id_vars=['Overall Average Crime Rate'], value_vars=religion_columns,
                           var_name='Religion', value_name='Percentage')

# Calculate the correlation between each religion and the overall average crime rate
correlation_data = long_format_data.groupby('Religion')['Overall Average Crime Rate'].corr(long_format_data['Percentage']).reset_index()

# Visualize the aggregated crime rates for each religion
fig = px.bar(correlation_data,
             x='Religion',
             y='Overall Average Crime Rate',
             title='Correlation between Religious Composition and Overall Average Crime Rate')
fig.show()
```
